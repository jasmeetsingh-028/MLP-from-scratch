```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Understanding derivatives

```python
def f(x):
    """A simple quadratic function."""
    return 3*x**2 - 2*x + 1
```

```python
f(3.0)
```

### plotting function

```python
x = np.linspace(-2,2, 100)  # Generate 100 points between -1 and 1
y = f(x)
plt.plot(x, y, 'r-', label='f(x) = 3x^2 - 2x + 1')
plt.title('Plot of the function f(x)') 
```

Derivative meaning in simple terms: how will the function f(a) change if we add a small value 'h' to it i.e.: f(a+h)

Calculating slop at h = 2.0

```python
h = 0.003 #small step size
# Calculate the derivative at a point a using the limit definition
a = 2
print('f(a) =', f(a))
print('f(a + h) =', f(a + h))
#obs: f(a+h) moves to the right
#plot f(a) and f(a + h) on the graph of f(x)
plt.plot(x, y, 'r-', label='f(x) = 3x^2 - 2x + 1')
plt.title('Plot of the function f(x)')
plt.plot(a, f(a), 'bo', label='f(a)')
plt.plot(a + h, f(a + h), 'ro', label='f(a + h)')
plt.legend()
```

To check how much f(a+h) is greater than f(a) we do f(a+h) - f(a)

```python
f(a+h) - f(a)  # This is the change in function value
# This is the change in x value
print('Change in function value:', f(a + h) - f(a))
```

```python
# to normalize the change in function value, we divide by h
print('Normalized change in function value or the SLOPE or df/dx at x = a = {}:'.format(a), (f(a + h) - f(a)) / h)
```

```python
def derivative_at_point(f, a, h):
    """Calculate the derivative of f at point a using the limit definition."""
    # print('f(a) =', f(a))
    # print('f(a + h) =', f(a + h))
    return (f(a + h) - f(a)) / h
```

## Plotting line of slope:


Using point slope equation:



the point on line of slope that we know is (x0, y0) = (a, f(a))

applying point slope equation to find y_intercept:

*y - f(a) = slope * (x - a)*

Since the point where line of slope cuts the y-axis (y_intercept) is the point where x = 0

*y_intercept - f(a) = slope * (-a)*

*y_intercept = f(a) - slope * a*

Equation of line of slope:

*y = slope * x + y_intercept*


```python
plt.plot(x, y, 'r-', label='f(x) = 3x^2 - 2x + 1')

#plot the tangent line at point a
tangent_slope = derivative_at_point(f, a, h)
tangent_y_intercept = f(a) - tangent_slope * a
plt.plot(x, tangent_slope * x + tangent_y_intercept, 'b--', label='Tangent line at a')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Tangent line at a = {}'.format(a))
plt.legend()
plt.grid()
```

Taking a more complex function

```python
a = 0.5
b = -3.0
c = 6.0

d = a*b+c
print('d = a*b + c = {}*{} + {} = {}'.format(a, b, c, d))
```

```python
## to calculate derivative of d with respect to a, we bump value of by h

## since a is bumped by h and b is a -ve value then (a+h) * b will be less than a*b
## so we expect the slope to be negative
h = 0.0003
print(f'd(a + h) = {(a+h) * b + c}')
print(f'd(a) = {d}')
print(f'slope = {(((a+h) * b + c) - d) / h}')

## SINCE B AND C ARE CONSTANT IF WE TAKE ## DERIVATIVE OF d WITH RESPECT TO a, WE GET b
print('Derivative of d with respect to a is:', b)
```

If we do this to b

```python
h = 0.0003
print(f'd(b + h) = {a * (b + h) + c}')
print(f'd(b) = {d}')
print(f'slope = {((a * (b + h) + c) - d) / h}')

## SINCE A AND C ARE CONSTANT IF WE TAKE ## DERIVATIVE OF d WITH RESPECT TO b, WE GET A
print('Derivative of d with respect to b is:', a)
```
