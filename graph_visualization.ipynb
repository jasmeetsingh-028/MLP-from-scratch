```python
from graphviz import Digraph
```

```python
##draw simple graph with graphviz
dot = Digraph(comment='Simple Graph')
dot.node('A', 'Node A')
dot.node('B', 'Node B') 
dot.node('C', 'Node C')
dot.edges(['AB', 'AC'])  # Add edges between nodes A and B, and A and C
dot.edge('B', 'C', label='Edge from B to C')  # Add an edge from B to C with a ## display the graph in in the cell
dot
```

```python
##display the graph in the cell
dot.render('simple_graph', format='png', cleanup=True)  # Save the graph as a PNG file
dot.view()  # Open the graph in the default viewer
```

## Display simple linked list

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def append_node(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
            
    def display(self):  
        temp = self.head
        while temp:
            print(temp.value, end=' -> ')
            temp = temp.next
        print('None')
```

```python
LinkedList1 = LinkedList()
LinkedList1.append_node(1)  
LinkedList1.append_node(2)
LinkedList1.append_node(3)
LinkedList1.display()  # Output: 1 -> 2 -> 3 -> None
```

```python
## use graphviz to visualize the linked list
dot = Digraph(comment='Linked List')
current = LinkedList1.head
while current:
    dot.node(str(current.value), str(current.value))
    if current.next:
        dot.edge(str(current.value), str(current.next.value))
    current = current.next
## display the linked list graph
dot
```
